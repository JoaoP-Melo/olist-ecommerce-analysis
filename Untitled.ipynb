# Qual categoria vende mais por estado ? 

```python
order_items_analyses = order_items[["order_id", "product_id"]]
products_analyses = products[["product_id","product_category_name"]]
orders_analyses = orders[["order_id", "customer_id"]]
custumers_analyses = custumers[["customer_id", "customer_state"]]
```

Para encontrara a quantidade de vendas de cada categoria por estado tive que ruinir informaçoes de diversas tabelas, apos isso realizei um merge entre elas

```python
df_analyses = orders_analyses.merge(right=custumers_analyses, on="customer_id", )
df_analyses = df_analyses.merge(right= order_items_analyses, on = "order_id")
df_analyses = df_analyses.merge(right=products_analyses, on="product_id")
```

Apos fazer o merge entre elas agrupei por estado e categoria para somar a quantidade de vendas atravez de uma funcao que criei para retorna o maior valor das colunas que foram agrupadas

```python
df_analyses = (df_analyses[["customer_state", "product_category_name", "product_id"]]
              .groupby(["customer_state", "product_category_name"], as_index=False)
              .size())


def maior(row : pd.DataFrame):
    return row[['product_category_name', 'size']].sort_values('size', ascending=False).iloc[0]

df_analyses = df_analyses.groupby(["customer_state"], as_index=False).apply(maior)

df_analyses
```

Atravez dessa tabela e possivel notar um grande dominio da categoria beleza_saude no numero de vendas pelos estados do pais, alem disso  e possivel perceber que atras da categoria de beleza vem casa_mesa_banho que fica em primeiro no restante da maioria das regioes
