# Analise de Dados de Vendas (O-list Brasil)

## Objetivo
Analisar a distribuição de vendas no Brasil, identificando padrões regionais e categorias de maior desempenho, com o objetivo de gerar insights estratégicos para suporte à tomada de decisão.

## Dataset
O Dataset de dados importado tem varias colunas em diferentes tabelas foram utilizados merge entre elas, mas as colunas que foram usadas neste projeto sao:
- Id de indentificacao do pedido (order_id)
- Id de indentificacao do produto (product_id)
- Categoria do produto (product_category_name)
- Id de indentificacao do cliente (customer_id)
- Estado do cliente (customer_state)
- Preco do pedido (price)
- Nota dada pelo cliente ao pedido (review_score)
- Id de indentificacao do vendedor (seller_id)
- Data de aprovacao do pedido (order_approved_at)
- Data de saida do pedido ate o cliente (order_delivered_carrier_date)
- Data de entrega do pedido (order_delivered_customer_date)
- Data estimado para entrega do pedido (order_estimated_delivery_date)



# Importando Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Importando Dados

```python
customers = pd.read_csv("data\\olist_customers_dataset.csv")
orders = pd.read_csv("data\\olist_orders_dataset.csv")
order_items = pd.read_csv("data\\olist_order_items_dataset.csv")
order_payments = pd.read_csv("data\\olist_order_payments_dataset.csv")
order_review = pd.read_csv("data\\olist_order_reviews_dataset.csv")
products = pd.read_csv("data\\olist_products_dataset.csv")
sellers = pd.read_csv("data\\olist_sellers_dataset.csv")
category_name = pd.read_csv("data\\product_category_name_translation.csv")
```

# Perguntas de negocio

## Qual categoria vende mais por estado ? 

Objetivo: identificar quais categorias vendem mais em cada região do país.
Para responder esta pergunta foi necessario usar as informacoes order_id, product_id, product_category_name, customer_id, customer_state com essas informacoes foi possivel utilizar os dados para responder esta pergunta.

```python
order_items_analyses = order_items[["order_id", "product_id"]]
products_analyses = products[["product_id","product_category_name"]]
orders_analyses = orders[["order_id", "customer_id"]]
custumers_analyses = customers[["customer_id", "customer_state"]]
```

```python
categories_by_state = orders_analyses.merge(right=custumers_analyses, on="customer_id", )
categories_by_state = categories_by_state.merge(right= order_items_analyses, on = "order_id")
categories_by_state = categories_by_state.merge(right=products_analyses, on="product_id")
```

```python
categories_by_state = (categories_by_state[["customer_state", "product_category_name", "product_id"]]
              .groupby(["customer_state", "product_category_name"], as_index=False)
              .size()).rename(columns={"size" : "number_of_sales"})


def maior(row : pd.DataFrame):
    return row[['product_category_name', 'number_of_sales']].sort_values('number_of_sales', ascending=False).iloc[0]

categories_by_state = categories_by_state.groupby(["customer_state"], as_index=False).apply(maior)

categories_by_state = categories_by_state.sort_values('number_of_sales', ascending=False)

categories_by_state
```

Por meio dessa tabela, é possível notar um grande domínio da categoria beleza_saude no número de vendas pelos estados do país. Além disso, é possível perceber que, atrás da categoria de beleza, vem casa_mesa_banho, que fica em primeiro na das outras regiões. Isso mostra que as vendedores podem aumentar seus estoques de produtos de beleza, pois existe uma grande demanda no país.

```python
plt.figure(figsize=(8,5))
plt.barh(categories_by_state["customer_state"], categories_by_state["number_of_sales"])
plt.ylabel("Estado")
plt.xlabel("Quantidade de vendas")
plt.title("Numero de vendas da categoria lider de cada estado")
plt.gca().invert_yaxis()
```

Observa-se que os estados da região Sudeste apresentam um volume superior de vendas, mesmo quando considerando apenas a categoria mais vendida de cada estado, essa concentração pode estar associada ao fator de desenvolvimento da região.
O Sudeste representa o principal mercado consumidor e deve ser priorizado em estratégias logísticas e comerciais, regiões com menor volume podem representar oportunidades de crescimento e expansão de mercado.

## Quais categorias vendem mais no pais todo ?

Objetivo: encontrar as cinco categorias com maior nível de vendas no país todo. Para encontrar essa métrica, foram utilizadas as colunas order_id, product_id e product_category_name.

```python
order_items_analise = order_items[["order_id", "product_id"]]
products_analise = products[["product_id","product_category_name"]]

top_sales_categories = order_items_analise.merge(right=products_analise, on='product_id')

top_sales_categories = (top_sales_categories[["product_category_name", "order_id"]]
              .groupby("product_category_name", as_index=False)
              .size()).rename(columns={"size" : "number_of_sales"})


top_sales_categories["percent"] = top_sales_categories["number_of_sales"] / top_sales_categories["number_of_sales"].sum() * 100

top_sales_categories = top_sales_categories.sort_values("number_of_sales", ascending=False).iloc[0:5]

top_sales_categories
```

```python
plt.barh(top_sales_categories['product_category_name'], top_sales_categories['number_of_sales'])
plt.title("Categgorias com maior numero de vendas no pais")
plt.ylabel("Categorias")
plt.xlabel("Numero de vendas")
plt.gca().invert_yaxis()
```

É possível observar que não há uma distância discrepante de uma categoria para outra, na verdade, o número de vendas vai caindo de forma bem linear. É possível observar que apenas esse top cinco de categorias já representa aproximadamente 40% das vendas do país.
Essas categorias podem ser usadas como um incentivo em regiões com número de vendas menor, pois são as mais compradas no geral e poderiam aumentar o número de clientes nessas regiões com um mercado menor.

# Qual o ticket medio de cada estado ?

Objetivo: Analisar o volume de vendas por estado e encontrar o ticket médio Para obter essa métrica, foram utilizadas as colunas order_id, product_id, price, customer_id e customer_state.

```python
order_items_analise = order_items[["order_id", "product_id", "price"]]
orders_analise = orders[["order_id", "customer_id"]]
custumers_analise = customers[["customer_id", "customer_state"]]
```

```python
ticket_mean = order_items_analise.merge(right=orders_analise, on='order_id')
ticket_mean = ticket_mean.merge(right=custumers_analise, on='customer_id')
```

```python
ticket_mean = ticket_mean.groupby('customer_state', as_index=False).agg({'price' : 'median', 'customer_id': 'size'}).sort_values('customer_state')
ticket_mean = ticket_mean.rename(columns={'price' : 'mean', 'customer_id' : 'qty_sales'})
ticket_mean
```

A tabela mostra que as regiões do Sudeste brasileiro têm uma grande quantidade de vendas a mais que as outras regiões, assim, podemos analisar que o ticket médio delas é bem inferior ao das outras regiões.

```python
plt.scatter( ticket_mean['qty_sales'], ticket_mean['mean'])
plt.ylabel("Ticket medio")
plt.xlabel("Quantidade de vendas")
plt.title("Ticket Medio x Numero de Vendas")
```

Por meio da análise do gráfico, podemos comprovar que a relação entre ticket médio e valor tem uma tendencia de ser inversamente proporcional, quando um cresce, o outro decresce. Esse comportamento evidencia que estados mais desenvolvidos possuem uma atividade de vendas maior. Os vendedores poderiam colocar mais produtos à venda com o preço menor que o ticket médio das região com menor volume de vendas para tentar, de alguma forma, aumentar o número de vendas nela.

# Relacao entre valor e avaliacoes

Objetivo: Analisar se o valor da compra influencia na avaliacao da venda, para isso foi usado as colunas order_id, review_score, price.

```python
orders_review_analise = order_review[['order_id','review_score']]
order_items_analise = order_items[['order_id', 'price']]

price_and_evaluation = orders_review_analise.merge(right=order_items_analise, on='order_id')
price_and_evaluation = price_and_evaluation.groupby('order_id').agg({
    'price': 'sum',
    'review_score': 'first'
})
```

```python
np.corrcoef(price_and_evaluation['review_score'],price_and_evaluation['price'])
```

Após analisar a correlação do review com o preço, podemos perceber que é muito próxima de zero, indicando ausência de relação entre as métricas.

```python
price_and_evaluation[price_and_evaluation["price"] > 5000]

price_and_evaluation = price_and_evaluation[price_and_evaluation["price"] <= 5000]

plt.scatter(price_and_evaluation['price'], price_and_evaluation['review_score'])
plt.yticks(range(1, 6, 1))
plt.xlabel("Preço Total do Pedido")
plt.ylabel("Nota")
plt.title("Relação entre Preço e Avaliação")
```

Foram removidos pedidos com valores acima de R$ 5.000 por representarem outliers que poderiam atrapalhar a análise gráfica.
É possível perceber que, independentemente do preço, as avaliações estão bem distribuídas no gráfico. Para comprovar isso, podemos cortar o dataset em cinco faixas de preço e mostrar a média de avaliações em cada faixa.

```python
evaluation_analysis = orders_review_analise.merge(right=order_items_analise, on='order_id')
evaluation_analysis = evaluation_analysis.groupby('order_id').agg({
    'price': 'sum',
    'review_score': 'first'
})

evaluation_analysis = evaluation_analysis[evaluation_analysis["price"] <= 5000]

evaluation_analysis['faixa_preco'] = pd.qcut(evaluation_analysis['price'], 5)

evaluation_analysis.groupby('faixa_preco')['review_score'].mean()
```

Ao observar os pedidos em faixas de preço, existe leve tendência de redução na avaliação média conforme o valor do pedido aumenta, mas a diferença é pequena, indicando que o impacto do preço na satisfação do cliente é limitado.
A análise indica que o preço dos pedidos não apresenta relação significativa com as avaliações dos clientes, as avaliações permanecem consistentemente altas em todas as faixas de preço, sugerindo que outros fatores tenham mais impacto nas avaliações.

# Quais os dez vendedores com maior numero de vendas no pais

Objetivo: Buscar os dez vendedores que mais realizam vendas no país, usando as colunas product_id, seller_id e a tabela completa de sellers.

Por meio dessa tabela, é possível observar que os maiores vendedores são do estado de São Paulo, isso apenas confirma o alto nível de desenvolvimento da região e a consolida como número de vendas. Os outros vendedores poderiam usar estas informações para investir no mercado de outras regiões, pois é possível analisar que, apesar de São Paulo ter um número de clientes maior, também tem vários concorrentes. 

# Tempo médio de transporte em cada região

Objetivo: Buscar o numero de dias medio para o produto chegar na casa do cliente em cada estado, usando as colunas order_id customer_id, order_delivered_carrier_date,order_delivered_customer_date, customer_id e customer_state

```python
orders_analise  = orders[[ "order_id","customer_id", "order_delivered_carrier_date","order_delivered_customer_date"]]

custumers_analise = customers[["customer_id", "customer_state"]]

average_number_days = orders_analise.merge(right=custumers_analise, on="customer_id")

average_number_days = average_number_days.dropna()
```

Linhas que possuíam valores vazios foram retiradas da análise para não distorcer as métricas.

```python
def formata_data(data_completa : str):
    data_completa = str(data_completa)
    data = data_completa.split(' ')

    return data[0]

average_number_days["order_delivered_carrier_date"] = average_number_days["order_delivered_carrier_date"].apply(formata_data)
average_number_days["order_delivered_customer_date"] = average_number_days["order_delivered_customer_date"]. apply(formata_data)
```

```python
average_number_days["order_delivered_carrier_date"] = pd.to_datetime(average_number_days["order_delivered_carrier_date"])
average_number_days["order_delivered_customer_date"] = pd.to_datetime(average_number_days["order_delivered_customer_date"])
```

```python
average_number_days['total_delivery_time'] = average_number_days["order_delivered_customer_date"] - average_number_days["order_delivered_carrier_date"]

average_number_days['total_delivery_time'] = average_number_days['total_delivery_time'].astype(str)

def pega_dias(n_dias):
    n_dias = str(n_dias)
    dias = n_dias.split(' ')
    dias = int(dias[0])
    return dias

average_number_days['total_delivery_time'] = average_number_days['total_delivery_time'].apply(pega_dias)
```

```python
average_number_days = average_number_days.groupby('customer_state',as_index=False).agg({'total_delivery_time' : 'mean'}).sort_values('total_delivery_time')
```

```python
plt.figure(figsize=(8,5))
plt.barh(average_number_days["customer_state"], average_number_days["total_delivery_time"])
plt.ylabel("Estado")
plt.xlabel("Tempo medio de entregas")
plt.title("Tempo medio de entregas por regiao(Em dias)")
```

Observa-se que estados com maior volume de vendas apresentam menor tempo médio de entrega, esse padrão pode estar relacionado a uma infraestrutura logística mais eficiente nessas regiões.
Portanto, é possível observar que regiões com menor tempo de entrega podem oferecer melhor experiência ao cliente, e a logística pode influenciar o volume de compra.
