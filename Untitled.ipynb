# Esrados, tickets, categorias e numero de vendas

Para começar a analisar o dataset pensei em encontrar a quantidade de vendas por categoria

```python
order_items_analyses = order_items[["order_id", "product_id"]]
products_analyses = products[["product_id","product_category_name"]]
orders_analyses = orders[["order_id", "customer_id"]]
custumers_analyses = custumers[["customer_id", "customer_state"]]
```

Para encontrara a quantidade de vendas de cada categoria por estado tive que ruinir informaçoes de diversas tabelas, apos isso realizei um merge entre elas

```python
df_analyses = orders_analyses.merge(right=custumers_analyses, on="customer_id", )
df_analyses = df_analyses.merge(right= order_items_analyses, on = "order_id")
df_analyses = df_analyses.merge(right=products_analyses, on="product_id")
```

Apos fazer o merge entre elas agrupei por estado e categoria para somar a quantidade de vendas atravez de uma funcao que criei para retorna o maior valor das colunas que foram agrupadas

```python
df_analyses = (df_analyses[["customer_state", "product_category_name", "product_id"]]
              .groupby(["customer_state", "product_category_name"], as_index=False)
              .size())


def maior(row : pd.DataFrame):
    return row[['product_category_name', 'size']].sort_values('size', ascending=False).iloc[0]

df_analyses = df_analyses.groupby(["customer_state"], as_index=False).apply(maior)

df_analyses
```

Atravez dessa tabela e possivel notar um grande dominio da categoria beleza_saude no numero de vendas pelos estados do pais, alem disso  e possivel perceber que atras da categoria de beleza vem casa_mesa_banho que fica em primeiro no restante da maioria das regioes, iss omostra que as empresas podem aumentar seus estoque da categoria de beleza, pois existe uma grande demanda no pais


Com base no que foi observado ja e possivel saber que as categorias que mais vendem no pais, porem fica uma duvida sobre qual é a diferença de venda de cada categoria, para responder isso uni as tabelas novamente e as agrupei, mas dessa vez agrupei por categoria para pegar a informaçao do pais todo

```python
order_items_analise = order_items[["order_id", "product_id"]]
products_analise = products[["product_id","product_category_name"]]

df_analyses = order_items_analise.merge(right=products_analise, on='product_id')

df_analyses = (df_analyses[["product_category_name", "order_id"]]
              .groupby("product_category_name", as_index=False)
              .size())

df_analyses["percent"] = df_analyses["size"] / df_analyses["size"].sum() * 100

df_analyses = df_analyses.sort_values("size", ascending=False).iloc[0:5]

df_analyses
```

E possivel observar que nao ha uma distancia discrepante de uma categoria para outra, na verdade o numero de vendas vai decendo de forma bem linear, mas possivel observar que apenas esse top cinco categorias ja representa apoximadamente 40% das vendas do pais

Mas analisando essa porcentagem ficaria a duvida se alguma regiao esta inflando essa estatistica com um numero de compras muito superior as outra regioes

Para isso fiz a união das tabelas novamente agrupei os dados por estado e calculei com a função agg a media e a quantidade venda por estado

```python
order_items_analise = order_items[["order_id", "product_id", "price"]]
orders_analise = orders[["order_id", "customer_id"]]
custumers_analise = custumers[["customer_id", "customer_state"]]
```

```python
df_analyses = order_items_analise.merge(right=orders_analise, on='order_id')
df_analyses = df_analyses.merge(right=custumers_analise, on='customer_id')
```

```python
df_analyses = df_analyses.groupby('customer_state', as_index=False).agg({'price' : 'median', 'customer_id': 'size'}).sort_values('customer_state')
df_analyses = df_analyses.rename(columns={'price' : 'mean', 'customer_id' : 'qty_sales'})
df_analyses
```

Atravez dessa tabela é possivel observar que as regioes do sudeste brasileiro tem uma quantidade ezorbitante de vendas a mais que as outras regioes por isso podemos analisar que o ticket medio delas e bem inferior aos das outras regioes

# Avaliações, clientes, vendedores e tempo de entrega

Apos fazer as anlises envolvendos os estados e outros afins decidi olhar mais para os clientes para compraravar as metricas encontradas, para isso começei a analisar se o preço tinha influencia na avaliação

```python
orders_review_analise = order_review[['order_id','review_score']]
order_items_analise = order_items[['order_id', 'price']]

df_analyses = orders_review_analise.merge(right=order_items_analise, on='order_id')
df_analyses = df_analyses.groupby('order_id').agg({
    'price': 'sum',
    'review_score': 'first'
})
```

Começei definindo qual seria as colunas que usaria nesta analise

```python
np.corrcoef(df_analyses['review_score'],df_analyses['price'])
```

Apos usar o numpy para checar a correlação do review com o preço, podemos perceber que a relação entre eles e quase inexistente, mas decidi mesmo assim conferir e provar esta afirmação

```python
df_analyses[df_analyses["price"] > 5000]

df_analyses = df_analyses[df_analyses["price"] <= 5000]

plt.scatter(df_analyses['price'], df_analyses['review_score'])
plt.xlabel("Preço Total do Pedido")
plt.ylabel("Nota")
plt.title("Relação entre Preço e Avaliação")
plt.show()
```

Para comprovar a inforção comecei retirando alguns outliars que eram compras acima de 5 mill R$ que era uma quantidade insignificante para a analise, apos isso apenas decidid representar em forma de grafico a analíse

E possivel perceber que independente do preço as avalições estao bem concentradas e as unicas que estão distantes das outras e aquelas que possue um preço elevado comparado as outras compras

Para comprovar isso podemos cortar o dataset em cinco faixas de preço e mostrar a media de avaliações em cada faixa

```python
df_analyses = orders_review_analise.merge(right=order_items_analise, on='order_id')
df_analyses = df_analyses.groupby('order_id').agg({
    'price': 'sum',
    'review_score': 'first'
})

df_analyses = df_analyses[df_analyses["price"] <= 5000]

df_analyses['faixa_preco'] = pd.qcut(df_analyses['price'], 5)

df_analyses.groupby('faixa_preco')['review_score'].mean()
```

Olhando ainda para as avaliaçoes podemos verificar os dez vendedores que mais venderam produtos no pais, porem o dataset nao possue o nome do vendedor apenas o codigo dele

Para fazer isso agrupei os ids dos vendedores e somei quantidade de vendas de cada um e extrai os tres primeiros

```python
order_items_analise = order_items[["product_id","seller_id"]]
sellers_analise = sellers

df_analyses = order_items_analise.merge(right=sellers_analise, on="seller_id")
df_analyses = df_analyses.groupby("seller_id").agg({
    "product_id" : "size",
    "seller_state" : "first",
    "seller_zip_code_prefix" : "first"})

df_analyses["seller_zip_code_prefix"] = df_analyses["seller_zip_code_prefix"].astype("str")
df_analyses = df_analyses.sort_values("product_id", ascending=False).iloc[0:10]

plt.figure(figsize=(8,5))
plt.barh(df_analyses["seller_zip_code_prefix"], df_analyses["product_id"])
plt.ylabel("Zip vendendor")
plt.xlabel("Quantidade de Vendas")
plt.title("Quantidade de vendas de casa vendedor")
plt.gca().invert_yaxis()
```

Podemos tambem relacionada com entregas pegar o numero medio de dias para uma entrega ser realizada por estado

```python
orders_analise  = orders[["order_approved_at", "order_id","customer_id", "order_delivered_carrier_date","order_delivered_customer_date","order_estimated_delivery_date"]]

custumers_analise = custumers[["customer_id", "customer_state"]]

df_analyses = orders_analise.merge(right=custumers_analise, on="customer_id")

df_analyses = df_analyses[df_analyses["order_approved_at"].notnull()]
df_analyses['order_delivered_carrier_date'] = df_analyses['order_delivered_carrier_date'].fillna(df_analyses['order_approved_at'])
df_analyses['order_delivered_customer_date'] = df_analyses['order_delivered_customer_date'].fillna(df_analyses['order_estimated_delivery_date'])
```

Apos unir as tabelas fiz uma limpeza pois existia varios valores nulos nas colunas de aprovação de pedido nela tive que exluir os nulos pois era um valor insignificante, a coluna do dia que a entrega saiu quando o valor dela era nulo peguei o dia que o pedido era aprovado como referencia, e na coluna do dia que o pedido foi entregue quando era nulo subtitui pelo data limite que a entrega seria realizada

```python
def formata_data(data_completa : str):
    data_completa = str(data_completa)
    data = data_completa.split(' ')

    return data[0]

df_analyses["order_delivered_carrier_date"] = df_analyses["order_delivered_carrier_date"].apply(formata_data)
df_analyses["order_delivered_customer_date"] = df_analyses["order_delivered_customer_date"]. apply(formata_data)
```

Criei uma função que cortava o numero de horas minutos e segundos da data e deixava apenas os dias, meses e anos porem ela ainda deixava a coluna em formato string

```python
df_analyses["order_delivered_carrier_date"] = pd.to_datetime(df_analyses["order_delivered_carrier_date"])
df_analyses["order_delivered_customer_date"] = pd.to_datetime(df_analyses["order_delivered_customer_date"])
```

Utilizei uma função pandas para transformar em data

```python
df_analyses['tempo_de_entrega_total'] = df_analyses["order_delivered_customer_date"] - df_analyses["order_delivered_carrier_date"]

df_analyses['tempo_de_entrega_total'] = df_analyses['tempo_de_entrega_total'].astype(str)

def pega_dias(n_dias):
    n_dias = str(n_dias)
    dias = n_dias.split(' ')
    dias = int(dias[0])
    return dias

df_analyses['tempo_de_entrega_total'] = df_analyses['tempo_de_entrega_total'].apply(pega_dias)
```

Realizei o calculo de dias totais de frete em cada linha tranformei em str e joguei dentro de uma funcao que tirava tudo que nao era numero da coluna

```python
df_analyses = df_analyses.groupby('customer_state',as_index=False).agg({'tempo_de_entrega_total' : 'mean'})
```

No final agrupei por estado e fiz media de dias e por fim fiz um grafico

```python
plt.figure(figsize=(8,5))
plt.barh(df_analyses["customer_state"], df_analyses["tempo_de_entrega_total"])
plt.ylabel("customer_state")
plt.xlabel("Tempo medio de entregas")
plt.title("Tempo dmedio de entregas por regiao(Em dias)")
```
